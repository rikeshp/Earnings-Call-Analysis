```python
import pandas as pd
import PyPDF2 as pp3
```

```python
def get_info(path):
    with open(path, 'rb') as f:
        pdf = pp3.PdfFileReader(f)
        info = pdf.getDocumentInfo()
        number_of_pages = pdf.getNumPages()
    
    print(info)
    author = info.author
    creator = info.creator
    producer = info.producer
    subject = info.subject
    title = info.title
```

```python
get_info('letters/2p2_q1-2020.pdf')
```

```python
text_archive = []
def text_extractor(path):
    with open(path, 'rb') as f:
        pdf = pp3.PdfFileReader(f)
        # get the first page
        page = pdf.getPage(2)
        print(page)
        print('Page type: {}'.format(str(type(page))))
        text = page.extractText()
        text_archive.append(text)
        print(text)
```

```python
text_extractor('letters/2p2_q1-2020.pdf')
```

```python
import gensim as gs
import nltk
import re
```

```python
str1 = ' '.join(text_archive)
```

```python
str1 = re.sub('[^A-Za-z0-9]+', ' ', str1)
```

```python
str1
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize

text = str1
text_tokens = word_tokenize(text)

tokens_without_sw = [word for word in text_tokens if not word in stopwords.words()]

print(tokens_without_sw)
```

```python
# import these modules 
from nltk.stem import PorterStemmer 
from nltk.tokenize import word_tokenize 
   
ps = PorterStemmer() 
  
# choose some words to be stemmed 
words = tokens_without_sw

for w in words: 
    print(w, " : ", ps.stem(w)) 
```
