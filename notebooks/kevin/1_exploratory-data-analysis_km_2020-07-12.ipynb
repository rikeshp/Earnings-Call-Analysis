```python
import pandas as pd
import PyPDF2 as pp3
```

```python
def get_info(path):
    with open(path, 'rb') as f:
        pdf = pp3.PdfFileReader(f)
        info = pdf.getDocumentInfo()
        number_of_pages = pdf.getNumPages()
    
    print(info)
    author = info.author
    creator = info.creator
    producer = info.producer
    subject = info.subject
    title = info.title
```

```python
get_info('letters/2p2_q1-2020.pdf')
```

```python
text_archive = []
def text_extractor(path):
    with open(path, 'rb') as f:
        pdf = pp3.PdfFileReader(f)
        # get the first page
        page = pdf.getPage(2)
        print(page)
        print('Page type: {}'.format(str(type(page))))
        text = page.extractText()
        text_archive.append(text)
        print(text)
```

```python
text_extractor('letters/2p2_q1-2020.pdf')
```
